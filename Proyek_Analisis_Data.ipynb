## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

 - **Measurable**:Rcency diukur dalam hari sejak transaksi terakhir,Frequency diukur dari jumlah pesanan unik,Monetary diukur dari total nilai pembelian (R) setiap segmentasi menghasilkan angka kontribusi dan persentase.

  - **Time-bound**: Data yang dianalisis dibatasi dari 1 Agustus 2017 hingga 31 Juli 2018 (1 tahun fiskal). Ini mencakup siklus bisnis yang lebih panjang sehingga dapat mengidentifikasi pelanggan dengan repeat order.

### Cleaning Data

```python
# CLEANING DATA FINAL (OLIST E-COMMERCE)
# Konversi Data

date_cols_orders = ['order_purchase_timestamp', 'order_approved_at',
                    'order_delivered_carrier_date', 'order_delivered_customer_date',
                    'order_estimated_delivery_date']
for col in date_cols_orders:
    orders[col] = pd.to_datetime(orders[col], errors='coerce')

order_items['shipping_limit_date'] = pd.to_datetime(order_items['shipping_limit_date'], errors='coerce')
order_reviews['review_creation_date'] = pd.to_datetime(order_reviews['review_creation_date'], errors='coerce')
order_reviews['review_answer_timestamp'] = pd.to_datetime(order_reviews['review_answer_timestamp'], errors='coerce')

print("1. Konversi datetime selesai.")

# Penanganan Missing Value

# Kategori produk
missing_cat_before = products['product_category_name'].isnull().sum()
products['product_category_name'] = products['product_category_name'].fillna('unknown')
print(f"2a. Missing category products: {missing_cat_before} -> {products['product_category_name'].isnull().sum()}")

# 2b. Metadata produk (length, photos, ukuran, berat)
for col in ['product_name_lenght', 'product_description_lenght', 'product_photos_qty']:
    products[col] = products[col].fillna(0)
for col in ['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm']:
    median_val = products[col].median()
    products[col] = products[col].fillna(median_val)
print("2b. Missing metadata produk diisi (0 untuk length/photos, median untuk ukuran/berat).")


# Penanganan Outlier
# Outlier berat produk (<=0 atau >5000 gram) -> ganti median
median_weight = products['product_weight_g'].median()
outlier_mask = (products['product_weight_g'] <= 0) | (products['product_weight_g'] > 5000)
print(f"3. Outlier weight sebelum: {outlier_mask.sum()} baris")
products.loc[outlier_mask, 'product_weight_g'] = median_weight

# Outlier payment_value (<=0) -> ganti median dari payment_value positif
negative_payment = (order_payments['payment_value'] <= 0).sum()
if negative_payment > 0:
    median_pos = order_payments[order_payments['payment_value'] > 0]['payment_value'].median()
    order_payments.loc[order_payments['payment_value'] <= 0, 'payment_value'] = median_pos
    print(f"   {negative_payment} payment_value <=0 diganti dengan median ({median_pos}).")
else:
    print("   Tidak ada payment_value negatif atau nol.")


# Filter Pesanan Delivered (Wajib untuk analisis)
completed_orders = orders[orders['order_status'] == 'delivered'].copy()
completed_orders = completed_orders.dropna(subset=['order_delivered_customer_date', 'order_estimated_delivery_date'])
print(f"4. Pesanan delivered setelah cleaning: {len(completed_orders)} dari {len(orders)} total")


# Periode Analisi(TIME-BOUND)

# Pertanyaan 1: RFM (1 tahun: 1 Agustus 2017 - 31 Juli 2018)
start_date_rfm = '2017-08-01'
end_date_rfm = '2018-07-31'
mask_rfm = (completed_orders['order_purchase_timestamp'] >= start_date_rfm) & (completed_orders['order_purchase_timestamp'] <= end_date_rfm)
orders_rfm = completed_orders[mask_rfm].copy()
print(f"5a. Periode RFM ({start_date_rfm} s.d {end_date_rfm}): {len(orders_rfm)} pesanan")

# Pertanyaan 2: Delay & Review (Q1 2018: 1 Jan - 31 Mar 2018)
q1_start = '2018-01-01'
q1_end = '2018-04-01'   # eksklusif (hanya sampai 31 Maret)
mask_q1 = (completed_orders['order_purchase_timestamp'] >= q1_start) & (completed_orders['order_purchase_timestamp'] < q1_end)
orders_delay = completed_orders[mask_q1].copy()
print(f"5b. Periode delay (Q1 2018): {len(orders_delay)} pesanan")

# 6. Persiapan Data RFM (Pertanyaan 1)
# Agregasi payment per order (karena 1 order bisa memiliki >1 metode bayar)
payment_sum = order_payments.groupby('order_id')['payment_value'].sum().reset_index()
rfm_base = orders_rfm.merge(payment_sum, on='order_id', how='left')
rfm_base = rfm_base[['customer_id', 'order_delivered_customer_date', 'order_id', 'payment_value']].copy()
rfm_base = rfm_base.dropna(subset=['payment_value'])
rfm_base = rfm_base.drop_duplicates(subset=['order_id'])
print(f"6. Data RFM siap: {rfm_base.shape[0]} baris (pesanan unik)")

# Persiapan Data Delay & Riview (Pertanyaan 2)

products_clean = products.merge(category_translation, on='product_category_name', how='left')
products_clean['product_category_name_english'] = products_clean['product_category_name_english'].fillna('unknown')

delay_items = orders_delay.merge(order_items[['order_id', 'product_id']], on='order_id', how='inner')
delay_items = delay_items.merge(products_clean[['product_id', 'product_category_name_english']], on='product_id', how='left')
delay_items = delay_items.merge(order_reviews[['order_id', 'review_score']], on='order_id', how='left')
delay_items['delay_days'] = (delay_items['order_delivered_customer_date'] - delay_items['order_estimated_delivery_date']).dt.days
delay_clean = delay_items.dropna(subset=['review_score', 'delay_days'])
print(f"7. Data delay & review siap: {delay_clean.shape[0]} baris (item pesanan dengan review)")

# Simpan Data Bersih Ke Folder Dashboard

os.makedirs('./dashboard', exist_ok=True)
rfm_base.to_csv('./dashboard/main_data.csv', index=False)
delay_clean.to_csv('./dashboard/delay_analysis.csv', index=False)
print("8. Data bersih disimpan di folder ./dashboard/ (main_data.csv dan delay_analysis.csv)")


print("\n===== CLEANING DATA SELESAI =====")
print("Data siap untuk Exploratory Data Analysis (EDA).")
```

### Pertanyaan 1:

 Pertanyaan 1

Segmen pelanggan mana (berdasarkan nilai Recency, Frequency, Monetary) yang memberikan kontribusi tertinggi terhadap total pendapatan, dan bagaimana profil mereka dalam periode 1 tahun terakhir (Agustus 2017 – Juli 2018)?

```python

# ============================================
# EDA - PERTANYAAN 1: Segmentasi RFM (Final)
# ============================================

# Load data
rfm_df = pd.read_csv('./dashboard/main_data.csv')
rfm_df['order_delivered_customer_date'] = pd.to_datetime(rfm_df['order_delivered_customer_date'])

recent_date = rfm_df['order_delivered_customer_date'].max()
print(f"Tanggal transaksi terakhir: {recent_date.date()}")

# Agregasi per customer
rfm_agg = rfm_df.groupby('customer_id').agg({
    'order_delivered_customer_date': lambda x: (recent_date - x.max()).days,
    'order_id': 'nunique',
    'payment_value': 'sum'
}).reset_index()
rfm_agg.columns = ['customer_id', 'recency', 'frequency', 'monetary']

# Hapus data tidak valid
rfm_agg = rfm_agg[(rfm_agg['recency'] >= 0) & (rfm_agg['frequency'] > 0) & (rfm_agg['monetary'] > 0)]

# Fungsi untuk memberi skor 1-4 berdasarkan persentil (memaksa 4 kelompok)
def assign_score_by_quantile(column, reverse=False):
    # Gunakan pd.qcut dengan duplicates='drop' sebenarnya bisa gagal jika nilai unik < 4
    # Maka kita gunakan rank terlebih dahulu untuk memaksa data menjadi unik secara ordinal
    ranked = column.rank(method='first')
    # Bagi ranked menjadi 4 kelompok yang sama rata
    try:
        scores = pd.qcut(ranked, 4, labels=[1,2,3,4], duplicates='drop').astype(int)
    except:
        # Jika masih gagal, gunakan cut dengan bins berdasarkan persentil
        bins = np.percentile(ranked, [0, 25, 50, 75, 100])
        bins = np.unique(bins)
        scores = pd.cut(ranked, bins=bins, labels=[1,2,3,4][:len(bins)-1], include_lowest=True)
        scores = scores.astype(int)
    if reverse:
        scores = 5 - scores
    return scores

# Hitung skor
rfm_agg['r_score'] = assign_score_by_quantile(rfm_agg['recency'], reverse=True)   # recency kecil => skor besar
rfm_agg['f_score'] = assign_score_by_quantile(rfm_agg['frequency'], reverse=False)
rfm_agg['m_score'] = assign_score_by_quantile(rfm_agg['monetary'], reverse=False)

# RFM total score (3-12)
rfm_agg['rfm_score'] = rfm_agg['r_score'] + rfm_agg['f_score'] + rfm_agg['m_score']

# Segmentasi
def segment_rfm(score):
    if score >= 10:
        return 'Champions'
    elif score >= 8:
        return 'Loyal'
    elif score >= 6:
        return 'Potential'
    else:
        return 'At Risk'

rfm_agg['segment'] = rfm_agg['rfm_score'].apply(segment_rfm)

# Ringkasan per segmen
segment_summary = rfm_agg.groupby('segment').agg({
    'customer_id': 'count',
    'monetary': 'sum'
}).rename(columns={'customer_id': 'customer_count', 'monetary': 'total_revenue'})
segment_summary['revenue_percent'] = (segment_summary['total_revenue'] / segment_summary['total_revenue'].sum()) * 100
segment_summary['customer_percent'] = (segment_summary['customer_count'] / segment_summary['customer_count'].sum()) * 100
segment_summary = segment_summary.reindex(['Champions', 'Loyal', 'Potential', 'At Risk']).fillna(0)
print(segment_summary)

# Visualisasi 1: Kontribusi pendapatan
plt.figure(figsize=(10, 6))
colors_bar = ['#2E8B57', '#3CB371', '#66CDAA', '#FFA07A']
ax = sns.barplot(x=segment_summary.index, y='total_revenue', data=segment_summary.reset_index(), palette=colors_bar)
plt.title('Kontribusi Pendapatan per Segmen Pelanggan\n(Periode: 1 Agustus 2017 – 31 Juli 2018)', fontsize=14, fontweight='bold')
plt.xlabel('Segmen Pelanggan', fontsize=12)
plt.ylabel('Total Pendapatan (R$)', fontsize=12)
for i, v in enumerate(segment_summary['total_revenue']):
    ax.text(i, v + 500, f'R$ {v/1000:.0f}K', ha='center', fontsize=10)
plt.tight_layout()
plt.show()

# Visualisasi 2: Persentase pelanggan vs pendapatan
summary_plot = segment_summary[['customer_percent', 'revenue_percent']].reset_index()
summary_plot = summary_plot.melt(id_vars='segment', var_name='metric', value_name='percentage')
plt.figure(figsize=(10, 6))
sns.barplot(x='segment', y='percentage', hue='metric', data=summary_plot, palette=['#72BCD4', '#FF8C42'])
plt.title('Persentase Jumlah Pelanggan vs Kontribusi Pendapatan per Segmen', fontsize=14, fontweight='bold')
plt.xlabel('Segmen Pelanggan', fontsize=12)
plt.ylabel('Persentase (%)', fontsize=12)
plt.legend(title='Metrik', labels=['% Pelanggan', '% Pendapatan'])
for p in plt.gca().patches:
    plt.gca().annotate(f'{p.get_height():.0f}%', (p.get_x() + p.get_width()/2., p.get_height()),
                       ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.show()
```

**Insight dari Analisis RFM (Periode: 1 Agustus 2017 – 31 Juli 2018):**

1. **Konsentrasi Pendapatan pada Segmen Loyal & Champions**  
   Hanya 50% pelanggan (Champions + Loyal) memberikan kontribusi **68% dari total pendapatan**. Segmen Loyal sendiri menyumbang 40% pendapatan dengan jumlah pelanggan 34%, menjadikannya tulang punggung bisnis.

2. **Efisiensi Segmen Champions**  
   Champions (16% pelanggan) menghasilkan 28% pendapatan, artinya rata-rata pendapatan per pelanggan di segmen ini **Rp 278.000** (3,19 juta / 11.458), hampir 2x lipat dari rata-rata keseluruhan (total pendapatan 11,5 juta / 71.735 ≈ Rp 160.000). Mereka adalah aset utama yang perlu mendapatkan program loyalitas eksklusif.

3. **Potensi Besar dari Segmen Potential**  
   Sebanyak 34% pelanggan berada di segmen Potential dengan kontribusi pendapatan 26%. Mereka memiliki frekuensi dan recency sedang. Dengan strategi retensi yang tepat (misalnya rekomendasi produk personalisasi atau diskon terbatas), segmen ini dapat ditingkatkan menjadi Loyal atau bahkan Champions.

4. **Segmen At Risk: Tidak Signifikan namun Bukan Nol**  
   Hanya 16% pelanggan dengan kontribusi pendapatan 6%. Meskipun dampaknya kecil, mereka masih bisa direaktivasi dengan kampanye "We Miss You" yang murah (email + voucher kecil) karena biaya reaktivasi jauh lebih rendah daripada akuisisi pelanggan baru.
